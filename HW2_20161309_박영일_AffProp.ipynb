# `K Means`

```python
import numpy as np                              # 넘파이 배열과 다른 필요 기능을 쓰기위해
from sklearn import datasets                    # 데이터를 쉽게 불러올 수 있는 데이터 로딩 하는 것을 제공한다.

from sklearn.cluster import KMeans              # KMeans 를 하기 위해
from sklearn.cluster import AffinityPropagation # 친밀도 전파를 하기위해

import matplotlib.pyplot as plt                 # 그래프를 그리기 위해
from mpl_toolkits.mplot3d import Axes3D         # 결과를 3D 로 시각화하기위해
```

# `친밀도전파(Affinity Propagation) 알고리즘을 통해 IRIS 데이터셋을 군집화하기`

```python
# iris data loading
iris = datasets.load_iris()

X = iris.data
y = iris.target   # 학습을 하는 것이 아닌 군집의 결과가 올바른지 확인하기 위해서

# iris data set 에서 데이터 사이의 거리를 측정하여 최솟값, 메디안, 최댓값을 구하자. (X data 는 총 150 개이다.)

# Similarity Matrix 를 의미하는 리스트를 생성
sim_matrix = np.zeros((len(X),len(X)), np.float32)
for i in range (len(X)):
  main = X[i]
  for j in range(len(X)):
    # sii 는 고려하지 않으므로 그에 따른 분기문 처리
    if j != i:
      value = (-1)*pow(main-X[j],2)
      # 4 차원으로 출력되는 것을 하나로 합쳐준다.
      value = sum(value)
      sim_matrix[i][j] = value

min_value = np.min(sim_matrix)
median_value = np.median(sim_matrix)
max_value = np.max(sim_matrix)
print(f'최솟값 : {min_value} / 중앙값 : {median_value} / 최댓값 : {max_value}')
```

# 1. 친밀도 전파 - 최솟값

```python
# #############################################################################
# 친밀도 전파 알고리즘을 수행한다.

#[1] 자기유사도를 최솟값으로 설정한 경우,
af_min = AffinityPropagation(preference=min_value).fit(X)

cluster_centers_indices_min = af_min.cluster_centers_indices_
labels_affi_min = af_min.labels_

print(f'군집의 수 : {len(cluster_centers_indices_min)}')

# 결과를 시각화해보자
# 3차원 figure 를 그릴 준비를 한다.
fig_affi_min = plt.figure(1, figsize=(8, 6))    # (figure 넘버, width_height in inch)
# 3차원 시각화를 위한 준비
ax_affi_min = Axes3D(fig_affi_min, rect=[0,0,0.95,1], elev=0, azim=45)
# 3차원 상에 결과 그리기 (컬러코드 c 에 대표 인덱스(float) 를 대입하여 색을 부여한다.)
ax_affi_min.scatter(X[:,0], X[:,1], X[:,2], c=labels_affi_min.astype(float), edgecolor=None)      # 입력의 종류가 4가지 이므로 시각화가 불가능하므로 3개만 가져온다.
# 축별로 레이블을 진행하여 축의 의미를 부여한다.
ax_affi_min.w_xaxis.set_ticklabels([])
ax_affi_min.w_yaxis.set_ticklabels([])
ax_affi_min.w_zaxis.set_ticklabels([])

ax_affi_min.set_xlabel("0")
ax_affi_min.set_ylabel("1")
ax_affi_min.set_zlabel("2")

print(f'Kmeans Result\n{labels_kmeans_3}\n\nAffinity Propagation Result\n{labels_affi_min}') # Kmeans 알고리즘과의 유사도를 비교하자.
```
